```python
#plots

x = np.linspace(0,10,20)
print(x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x)
plt.plot(y)
plt.plot(z)
```

```python
fig = plt.figure()

ax = fig.add_axes(rect = [0,0,1,1])

print(type(ax))

ax.plot(x,x, label = 'x', c="r", ls = '--')
ax.plot(x,y, label = '2')
ax.plot(x,z, label = 'kvadratu')
ax.set_title("Grafikas")
ax.set_xlabel(" X ais", fontsize= "xx-large", rotation = 45)
ax.set_ylabel(" Y ais", fontsize= "xx-large", rotation = 45)
ax.legend()
```

```python
fig, axes = plt.subplots(2,3 ) #2 eiles ir 3 stulpeliai
axes [1,2]. plot(x,z)
axes [0,0]. plot(x,x)

for row in axes:
    for col_ax in row:
        col_ax.plot(x,y,c="r")
```
